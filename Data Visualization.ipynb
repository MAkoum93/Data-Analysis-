```python
def sortColumn(data, col):
    data.sort_values(by=[col],inplace=True)
    print(df[col])

sortColumn(df,'state')

sortColumn(df,'locality')

# sorting a column in ascending order
    
```

```python
def sortMultiColumns(data, col1,col2,col3):
    data.sort_values(by=[col1,col2,col3],inplace=True)
    print(df[[col1,col2,col3]])

sortMultiColumns(df,'mo','dy','yr')
# sorting multiple columns in ascending order
```

```python
# function to change column values to numeric data type
def columnNum(data,col):
    mac=pd.Series(data[col])
    pd.to_numeric(mac,errors='ignore')
    print(mac)
    
print(columnNum(df,'yr') ) 
print(columnNum(df,'dy'))
```

```python
def scatterPlot(df):
    grid = sb.PairGrid( data = df,vars = list(df),height = 4)
    for x, y in zip(*np.triu_indices_from(grid.axes, k = 0)):
        grid.axes[x,y].set_visible(False)
    grid = grid.map_lower(plt.scatter, s = 0.4)
    grid = grid.map_upper(plt.scatter, s = 0.4)
    grid.set(alpha = 1)
    return grid
plt.rcParams["axes.labelsize"] = 14
scatterPlot(df[['yr','mo','dy','period','recordID','stake']])

 # function to produce a scatter plot Matrix over all numeric data type columns in the data set   
```
